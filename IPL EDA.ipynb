```python
# Importing pre-requisite packages...

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Extracting the csv file from PC

data = pd.read_csv("C:/Users/ripoo/OneDrive/Documents/IPL dataset/cricket_data.csv")
```

```python
# Describing the Data set

data.describe()
```

```python
# Checking for any missing values

print(data.isnull().sum())
```

```python
# Step 1 : Convert Data types

numeric_cols = ['Year', 'Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score',
                'Batting_Average', 'Balls_Faced', 'Batting_Strike_Rate', 'Centuries', 'Half_Centuries',
                'Fours', 'Sixes', 'Catches_Taken', 'Stumpings', 'Matches_Bowled',
                'Balls_Bowled', 'Runs_Conceded', 'Wickets_Taken', 'Best_Bowling_Match', 'Bowling_Average', 
                'Economy_Rate', 'Bowling_Strike_Rate', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls']

data[numeric_cols] = data[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# Select only numeric columns

numeric_columns = data.select_dtypes(include='number')
```

```python
# Fill missing values in numeric columns with the mean of each column

numeric_columns.fillna(numeric_columns.mean(), inplace=True)
```

```python
# Display the DataFrame with filled missing values

print(numeric_columns)
```

```python
# Step 3 : Explore Data 

summary_stats = data.describe()
print("Summary Statistics: \n", summary_stats)
```

```python
# Convert non-numeric columns to numeric ones if possible
numeric_data = data.apply(pd.to_numeric, errors='coerce')

# Compute correlation matrix
correlation_matrix = numeric_data.corr()

# Print correlation matrix
print("Correlation Matrix:\n", correlation_matrix)
```

```python
# Step 4 : Visualize the Data

plt.figure(figsize=(12, 8))
sns.histplot(data=data, x = 'Runs_Scored', bins = 20, kde = True)
plt.title('Distribution of Runs Scored by Players')
plt.xlabel('Runs_Scored')
plt.ylabel('Frequency')
plt.show()
```

```python
# Step 5 : Further anlysis

top_runs_scorers = data.sort_values(by='Runs_Scored', ascending=False).head(10)
print("Top Run Scorers:\n", top_runs_scorers)
```

```python
top_wickets_taker = data.sort_values(by='Wickets_Taken', ascending=False).head(10)
print(top_wickets_taker)
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data = data, x = 'Year', y = 'Runs_Scored')
plt.title('Trend of Runs Scored Over Years')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.show()
```

```python
# Example: Histogram for Runs_Scored
sns.histplot(data = data, x = 'Runs_Scored',  bins=20)
plt.xlabel('Runs_Scored')
plt.ylabel('Frequency')
plt.title('Histogram of Runs Scored')
plt.show()
```

```python
# Example: Scatter plot between Batting_Average and Runs_Scored

sns.scatterplot(data = data, x = 'Batting_Average', y = 'Runs_Scored')
plt.title('Scatter plot b/w Batting Average and Runs Scored')
plt.xlabel('Batting_Average')
plt.ylabel('Runs_Scored')
plt.show()
```

```python
# Exclude non-numeric columns before computing correlation
numeric_data = data.select_dtypes(include=['int', 'float'])

# Compute Correlation Matrix
correlation_matrix = numeric_data.corr()

# Create a heatmap
plt.figure(figsize=(20, 16))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f',  cmap='coolwarm', linewidths='.5')
plt.title('Correlation Matrix of IPL Data')
plt.show()
```

```python
# Sort the DataFrame by 'Runs_Scored' column in descending order

top_ten_players = data.sort_values(by='Runs_Scored', ascending=False).head(10)
print(top_runs_scorers)

plt.figure(figsize=(12, 8))
sns.barplot(data = top_ten_players, x = 'Player_Name', y = 'Runs_Scored')
plt.title('Top Players by Runs_Scored')
plt.xticks(rotation=45, ha='right') # Imp
plt.show()
```

```python
# Select the top ten players with the highest scores

highest_score_players = data.sort_values(by='Highest_Score', ascending=False).head(10)

# Visualize Highest Score by the top ten players

plt.figure(figsize=(10, 8))
sns.barplot(data = highest_score_players, x = 'Player_Name', y = 'Highest_Score', linewidth=1, edgecolor = 'black', palette="magma")
plt.title('\nTop Ten Players by Highest_Score \n \n')
plt.xticks(rotation=45, ha='right') # Imp
plt.show()

```

```python
# Convert 'Half_Centuries' column to numeric type

data['Half_Centuries'] = pd.to_numeric(data['Half_Centuries'], errors='coerce')

# Select the top ten players with the most half centuries

Most_Half_Centuries = data.sort_values(by = 'Half_Centuries', ascending=False).head(10)

# Visualize Half Centuries by the top ten players

plt.figure(figsize=(12, 8))
sns.barplot(data = Most_Half_Centuries, x = 'Player_Name', y = 'Half_Centuries', linewidth=1)
plt.title('\nTop Ten Players by Most Half Centuries \n \n')
plt.xticks(rotation=45, ha='right') # Imp
plt.show()

```

```python
# Group the data by 'Player_Name' and count occurrences
player_appearances = data.groupby('Player_Name')['Year'].count().sort_values(ascending=False)

# Select the top ten players with the most appearances
top_ten_players = player_appearances.head(10)

# Visualize the top ten players with the most appearances
plt.figure(figsize=(12, 6))
plt.bar(top_ten_players.index, top_ten_players.values, color='skyblue')
plt.xlabel('Player Name')
plt.ylabel('Number of Appearances')
plt.title('Top Ten Players with Most Appearances')
plt.xticks(rotation=45, ha='right')  # Adjust rotation for better readability
plt.show()
```

```python
# Convert 'Fours' and 'Sixes' columns to numeric type
data['Fours'] = pd.to_numeric(data['Fours'], errors='coerce')
data['Sixes'] = pd.to_numeric(data['Sixes'], errors='coerce')


# Sort the DataFrame by 'Fours' column in descending order
most_fours = data.sort_values(by = 'Fours', ascending = False).head(10)


# Visualize Fours by the top ten players
plt.figure(figsize=(10, 6))
sns.barplot(data =  most_fours, x = 'Player_Name' , y = 'Fours', linewidth=1, color='royalblue')
plt.title('\n Top Ten Players by Number of Fours Scored \n \n')
plt.xticks(rotation=45, ha='right') # Imp
plt.show()


# Sort the DataFrame by 'Sixes' column in descending order
most_sixes = data.sort_values(by = 'Sixes', ascending = False).head(10)


# Visualize Sixes by the top ten players
plt.figure(figsize=(10, 6))
sns.barplot(data =  most_sixes, x = 'Player_Name' , y = 'Sixes', linewidth=1, color='orange')
plt.title('\n Top Ten Players by Number of Sixes Scored \n \n')
plt.xticks(rotation=45, ha='right') # Imp
plt.show()



```
